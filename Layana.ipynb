```python
def deposit(bal):
  deposit_amount = float(input("enter your amount"))
  bal += deposit_amount
  print("current value = ",bal)
  return bal
def withdrawal(bal):
  withdrawal_amount = float(input("enter the amount to be withdrawn"))
  bal -= withdrawal_amount
  print("current balance =",bal)
  return bal
def balance(bal):
  print("current balance",bal)
def main():
  bal = 200
  while True:
    print("========MENU========")
    print("\n 1. Deposit")
    print("\n 2. Withdraw")
    print("\n 3. View Account Balance")
    print("\n 4. Exit")
    user_choice = input("Enter your choice: ")
    if user_choice =='1':
      bal = deposit(bal)
    elif user_choice == '2':
      bal = withdrawal(bal)
    elif(user_choice == '3'):
      balance(bal)
    else:
      break


main()
```
